## **1) ENCRYPTION USING HILL CIPHER**

```python
keyMatrix = [[0] * 3 for i in range(3)]

# Generate vector for the message
messageVector = [[0] for i in range(3)]
# Generate vector for the cipher
cipherMatrix = [[0] for i in range(3)]
# Following function generates the
# key matrix for the key string
def getKeyMatrix(key):
  k = 0
  for i in range(3):
    for j in range(3):
      keyMatrix[i][j] = ord(key[k]) % 97
      k += 1

# Following function encrypts the message
def encrypt(messageVector):
  for i in range(3):
    for j in range(1):
      cipherMatrix[i][j] = 0
      for x in range(3):
        cipherMatrix[i][j] += (keyMatrix[i][x] *messageVector[x][j])
        cipherMatrix[i][j] = cipherMatrix[i][j] % 26

def HillCipher(message, key):
# Get key matrix from the key string
  getKeyMatrix(key)
# Generate vector for the message
  for i in range(3):
    messageVector[i][0] = ord(message[i]) % 97

# Following function generates
# the encrypted vector
    encrypt(messageVector)

# Generate the encrypted text
# from the encrypted vector
    CipherText = []
    encrypted_text= []
    for i in range(3):
      CipherText.append(chr(cipherMatrix[i][0] + 97))

# Finally print the ciphertext
      Ciphertext= ("".join(CipherText))
    #print(Ciphertext)
    return Ciphertext




# Get the message to
# be encrypted
message = open("/content/Three Men in a Boat (1).txt","r+")
data = message.read()
x=3
input_text = data.lower()
res=[input_text[y-x:y] for y in range(x, len(input_text)+x,x)]
input=''.join(res)
print("INPUT TEXT: ",input)




# Get the key
key = "GYBNQKURP"
encrypted_text=[]
for i in range(0,len(res)):
  output=HillCipher(res[i], key)
  encrypted_text.append(output)
#print(encrypted_text)
result1 = ''.join(encrypted_text)
print("ENCRYPTED TEXT: " ,result1)




```

## **CONVERSION TO BINARY FORMAT**

```python
def convertion(text):
  ascii_value =[]
  binary_str=[]

  for i in range(len(text)):
    ascii_value.append(ord(text[i]))

  for i in range(len(ascii_value)):
    binary_value = bin(ascii_value[i])[2:]
    while len(binary_value) < 8:
      binary_value += '0'
      bnr = binary_value[::-1]
      binary_str.append(bnr)
  return binary_str

binary_ciphertext_1=convertion(result1)
print("CIPHER 1 :",binary_ciphertext_1)
binary_ciphertext_2=convertion(result2)
print("CIPHER 2 :",binary_ciphertext_2)
binary_plaintext=convertion(input)
print("INPUT : ",binary_plaintext)
```

```python
def spliting_values(text):
  row = []
  for i in range(len(text)):
    if len(text[i]) < 8:
      pass
    else:
      for j in range(0,8):
        row.append(text[i][j])
  return row

cipher_row_1 = spliting_values(binary_ciphertext_1)
cipher_row_2 = spliting_values(binary_ciphertext_2)
input_text = spliting_values(binary_plaintext)
```
